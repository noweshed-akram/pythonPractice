```python
import scipy.signal
```

```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
img=cv2.imread("test.jpg")
```

```python
gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_img,cmap="gray")
```

```python
gray_img.shape
```

```python
#Steps:
#Prepare a 20X20 pixel filter
filter_h=np.zeros((20,20))
```

```python
filter_h
```

```python
# Put the data on the filter_h using gaussian distribution
sigma=20
for x in range(20):
    for y in range(20):
        filter_h[x,y]=np.exp(-((x)**2+(y)**2)/(2*np.pi*(sigma)**2))
```

```python
filter_h
```

```python
plt.imshow(filter_h,cmap="gray")
```

```python
plt.imshow(gray_img,cmap="gray")
```

```python
gray_blur=scipy.signal.convolve2d(gray_img,filter_h,mode="same")
```

```python
plt.imshow(gray_blur,cmap="gray")
```

```python
gray_img_negative=gray_img - gray_blur
```

```python
plt.imshow(gray_img_negative,cmap="gray")
```

```python
gray_img_new_2=gray_img+0.00005*gray_img_negative
```

```python
plt.imshow(gray_img_new_2,cmap="gray")
```
